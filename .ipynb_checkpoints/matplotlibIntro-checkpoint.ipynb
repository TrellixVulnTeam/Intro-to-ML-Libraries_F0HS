## Plotting from pandas Dataframes

```python
import pandas as pd
```

```python
# Make a dataframe
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','',regex="true")
car_sales
```

```python
# Remove last two zeroes
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sales Date"]=pd.date_range("1/1/2022",periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Plot the total sales
car_sales.plot(x="Sales Date",y="Total Sales");
```

```python
car_sales["Price"]=car_sales["Price"].astype(int)
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter");
```

```python
# Another Dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Create histogram of age
heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#PyPlot method
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target');
```

```python
# OO method mixed with PyPlot
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
## OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

## Plot the data
scatter= ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

## Customize the plot
ax.set(title="Heart Disease and Cholestero Levels",
      xlabel="Age",
      ylabel="Cholesterol")

## Add  a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a hline
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
over_50.head()
```

```python
# Subplot of chol,age,thalach
fig, (ax0,ax1)= plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,6),
#                             constrained_layout=True,
                            sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(title="Heart disease and cholesterol levels",
       label="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");
# Add data to ax1
scatter= ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add meanline
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
#Set the style
plt.style.use('seaborn-whitegrid')

## OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

## Plot the data
scatter= ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter") # this changes the color scheme

## Customize the plot
ax.set(title="Heart Disease and Cholestero Levels",
      xlabel="Age",
      ylabel="Cholesterol")

## Add  a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a hline
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
# Subplot of chol,age,thalach
fig, (ax0,ax1)= plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,8),
#                             constrained_layout=True,
                            sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")
# Setting the x limit
ax1.set_xlim(50,80)

# Customize ax0
ax0.set(title="Heart disease and cholesterol levels",
       label="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");
# Add data to ax1
scatter= ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter")
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add meanline
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold")

fig.savefig("images/heart-disease-analysis-plot.png");

```
